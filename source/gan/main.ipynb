```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
image = Image.open('/usr/src/data/images/0000.jpg')
plt.imshow(image)
```

```python
from model import Discriminator, Generator
from utils import train
from loader import get_dataset
```

```python
# consts
latent_dim = 200
img_shape = (3, 64, 64)
img_size = (img_shape[1], img_shape[2])

use_leaky = False
normalize = True
lr = 2.e-4
beta1 = 0.5
beta2 = 0.999

batch_size = 32
epochs = 300
sample_interval = 500

# data
dataset = get_dataset(image_size=img_size, batch_size=batch_size)

# model
G = Generator(latent_dim=latent_dim, img_shape=img_shape, use_leaky=use_leaky, normalize=normalize)
D = Discriminator(img_shape=img_shape, use_leaky=True, normalize=normalize)
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
G.to(device)
D.to(device)

# optimizer
G_optim = optim.Adam(G.parameters(), lr=lr, betas=(beta1, beta2))
D_optim = optim.Adam(D.parameters(), lr=lr, betas=(beta1, beta2))

# loss
criterion = nn.BCELoss()
```

```python
train(
    dataset=dataset,
    epochs=epochs,
    latent_dim=latent_dim,
    G=G,
    G_optim=G_optim,
    D=D,
    D_optim=D_optim,
    criterion=criterion,
    device=device,
    sample_interval=sample_interval
)
```
